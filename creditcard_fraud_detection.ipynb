```python
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
```

```python
X=dt.iloc[:,:-1]
y=dt['Class']



#this function is used to remove the last column i.e class
```

```python
X_train, X_test, y_train, y_test= train_test_split(X , y , test_size=0.35)
```

x_train, y_train , x_test, y_test is udes to split the data into training and test data<br>
whereas<br>
(x_train ,y_train), (x_test, y_test) is used to load the already divided data into test and train dataset

```python
clf= linear_model.LogisticRegression(C=1e5)
```

1e5 is a number expressed using scientific notation and it means 10 to the 5th power (the e meaning 'exponent') so 1e5 is equal to 100000 , both notations are interchangeably meaning the same.<br>
<br>
C in linear_model.LogisticRegression(C=1e5) means:- <br>
The C parameter controls the amount of regularization in the LogisticRegression object: a large value for C results in less regularization<br>
Regularization is used to avoid overfitting of the model

```python
clf.fit(X_train, y_train)
```

```python
y_predict= np.array(clf.predict(X_test))
y_test =np.array(y_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
c=confusion_matrix(y_test, y_predict)
c
```

```python
a=accuracy_score(y_test, y_predict)
a
```

```python
r=classification_report(y_test, y_predict)
r
```
